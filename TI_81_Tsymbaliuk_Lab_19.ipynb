За допомогою $[x_{ij}]_{i = 1,2 \textrm{; } j = 1,...,n}$ - дійсної матриці на площині задано n точок так, що $x_{1j}, x_{2j}$ - координати j-тої точки. Точки попарно з'єднані відрізками. Знайти довжину найбільшого відрізка.

```python
import random
import matplotlib.pyplot as plt
from math import sqrt


def main():
    try:
        n = int(input('Enter n: '))
    except ValueError:
        print('Value error')
        return
    array = [(round(random.uniform(1, 40), 3), round(random.uniform(1, 40), 3)) for i in range(n)]
    print(array)
    plt.scatter([elem[0] for elem in array], [elem[1] for elem in array])
    for i in range(0, n-1, 2):
        plt.plot([array[i][0], array[i+1][0]], [array[i][1], array[i+1][1]])
    plt.show()
    print('Max:', max([sqrt((array[i+1][0]-array[i][0])**2+(array[i+1][1]-array[i][1])**2) for i in range(0, n-1, 2)]))


if __name__ == "__main__":
    main()
```
