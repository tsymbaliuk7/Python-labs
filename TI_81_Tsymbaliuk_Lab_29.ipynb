Дано дійсні числа $x_1, y_1, x_2, y_2, ..., x_n,y_n$.
 Відомо, що точки $p_1, p_2, ..., p_n$ з координатами $(x_1, y_1), (x_2, y_2),...,(x_n, y_n)$ попарно різні. Розглянемо
  замкнуту ламану $p_1p_2...р_n$. 
  
  а) Чи правильно, що ламана не має самоперетинів? 
  
  б) Припускаючи , що ламана не має самоперетинів, з'ясувати, чи є $n$-кутник $p_1p_2. . .p_n$ опуклим.

```python
import matplotlib.pyplot as plt
from sympy.geometry import *
from random import randint

def main():
    n=5
    points = [Point(randint(-30, 30), randint(-30, 30)) for i in range(n)]
    points.append(points[0])
    x = [i[0] for i in points]
    x.append(points[0][0])
    y = [i[1] for i in points]
    y.append(points[0][1])
    plt.plot(x,y)
    sides = [Segment2D(points[i],points[i+1]) for i in range(0, len(points) - 1)]
    res = [[intersection(sides[i], sides[j]) for j in range(len(sides)) if i != j] for i in range(len(sides))]
    res = [elem.count([])==2 for elem in res].count(False) != 0
    print('a) Broken line has a self-intersection:',res)
    figure = Polygon(points[0], points[1], points[2], points[3], points[4])
    print('b) Figure is convex:', figure.is_convex())
        
if __name__ == '__main__':
    main()
```
