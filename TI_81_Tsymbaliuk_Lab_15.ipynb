Дано натуральні числа $ x_1, y_1, r_i, х_2, у_2, r_2, ... , x_n, y_n, r_n.$ Числа $ x_i, y_i $  є центрами кіл радіуса$ r_i .$ Побудувати на екрані кола, задані послідовністю $ x_1, y_1, r_i, х_2, у_2, r_2, ... , x_n, y_n, r_n,$ а потім зафарбувати їх (одним і тим же кольором або різними кольорами). Процес побудови повинен починатися кругом з номером 1 і закінчуватися кругом з номером n; процес зафарбовування повинен відбуватися в зворотньому порядку - починатися кругом з номером n і закінчуватися кругом з номером 1.

```python

import random
import matplotlib.pyplot as plt

n = int(input('Enter n: '))
array = random.sample(range(1, 70), 3 * n)
print(array)
plt.axes()
for i in range(0, len(array), 3):
    plt.gca().add_patch(plt.Circle((array[i], array[i+1]), array[i+2], color='r', fill=False))
plt.axis('scaled')
plt.show()
for i in range(0, len(array), 3):
    plt.gca().add_patch(plt.Circle((array[len(array)-i-3], array[len(array)-i-2]), array[len(array)-i-1], color='r', fill=True))
plt.axis('scaled')
plt.show()
```
